```python
import holoviews as hv

data = df[df["Dataset"] == "Extrapolation"]

obs = hv.Scatter(
    data,
    kdims="L (real)",
    vdims="Obs (x1e10)",
    label="Observed"
).opts(
    color="red",
    marker="s",
    size=6
)

pred = hv.Scatter(
    data,
    kdims="L (real)",
    vdims="Model",
    label="Predicted"
).opts(
    color="royalblue",
    marker="circle",
    size=6
)

fig = (obs * pred).opts(
    width=800,
    height=600,
    # logx=True,
    logy=True,
    xlabel="Luminosity, L",
    ylabel="Schechter Function (10⁻¹⁰ φ(L))",
    title="Extrapolation (Observed vs Predicted)",
    show_legend=True,
    legend_position="top_right",
    fontsize={"title": 14,"labels": 12,"ticks": 12,"legend": 12},
    xformatter='%.0e'
    # xlim=(1e5,1e8)

)
fig
```

```python
hv.save(fig,"Schechter_extrapolation.html")
```

```python
import random
import plotly.graph_objects as go
import plotly.express as px
```

```python
style_map = {
    "Validation": {"size": 8, "color": "red", "symbol": "diamond"},
    "Extrapolation": {"size": 5, "color": "green", "symbol": "circle"},
    "Training": {"size": 5, "color": "blue", "symbol": "square"}
}



fig = go.Figure()

# Add each group to the chart one by one
for species, styles in style_map.items():
    df_sub = df[df["Dataset"] == species]
    # text_labels = [str(round(df_sub.loc[i, 'MAE']/df_sub.loc[i, 'Obs (x1e10)']*100,1)) if df_sub.loc[i, 'MAE'] >= 1 else "" for i in df_sub.index]
    text_labels = [str(round(df_sub.loc[i, 'MAPE'],1)) if df_sub.loc[i, 'MAPE'] >= 10 else "" for i in df_sub.index]
    text_positions = [random.choice(['top center','bottom center']) for _ in text_labels]
    print(text_labels)
    #list(df_sub['MAE'].where(df_sub['MAE'] >= 1, "").values)
    fig.add_trace(go.Scatter(
        x=df_sub["L (real)"],
        y=df_sub["Obs (x1e10)"],
        mode="markers+text", 
        name=species,
        # text=text_labels,
        # textposition=text_positions,
        marker=dict(
            size=styles["size"],
            color=styles["color"],
            symbol=styles["symbol"]
        )
    ))
fig.update_xaxes(type="log",exponentformat="power")
fig.update_yaxes(type="log",exponentformat="power")

fig.update_layout(
    xaxis=dict(title="Luminosity, L"),
    yaxis=dict(title="Schechter Function, 10<sup>-10</sup>φ(L)"),
    # title="Schechter",
    legend=dict(orientation="h", x=0.02,y=0.15),
    template='simple_white',
    width=800,
    height=600,
    legend_font_size=18,         # Legend items
    
)
fig.update_xaxes(
    tickfont=dict(size=14, family="Times New Roman"),
    title_font=dict(size=16)
)

fig.update_yaxes(
    tickfont=dict(size=14, family="Times New Roman"),
    title_font=dict(size=16)
)

fig.show()
```
