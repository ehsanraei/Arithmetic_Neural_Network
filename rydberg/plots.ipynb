```python
import holoviews as hv
from holoviews import dim
import pandas as pd
import numpy as np
import random
```

```python
hv.extension('bokeh')
```

```python
df = pd.read_excel("./rydberg_performance.xlsx","seed8_rmsprop_0.001")
```

```python
df['obs_round'] = df['Obs'].round(3)
df['model_round'] = df['Model'].round(3)
df['MAPE'] = round((df['Obs']-df['Model'])/df['Obs']*100,2).abs()
# df['yoffset_'] = 0.2
# df['n1_pos'] = df['n1'].apply(lambda x:x*(1+random.randrange(-10,10)/100))
# df['n2_pos'] = df['n2'].apply(lambda x:x*(1+random.randrange(-10,10)/100))
```

```python
df['MAPE'] = df.apply(lambda x:str(x['MAPE'])+'%' if (x['n2']-x['n1'])==1 else str(x['MAPE']),axis=1)
```

```python
fig = hv.Scatter(df,kdims=['n1'],vdims=['n2','Dataset']).opts(width=500,height=500,color=dim('Dataset').categorize({'Training': 'blue','Validation':'red','Extrapolation': 'green'}) \
                                                        ,size=10,marker=dim('Dataset').categorize({'Training': 'square','Validation':'diamond','Extrapolation': 'circle'}),legend_position='bottom_right')

fig = fig*hv.Labels(df,kdims=['n1','n2'],vdims=['MAPE']).opts(text_font_size='10pt',yoffset=0.3)
# fig = fig*hv.Labels(df,kdims=['n1','n2'],vdims=['model_round','yoffset_']).opts(text_font_size='10pt',yoffset=-0.3)


fig.opts(xticks=8,yticks=10,show_legend=True)
```

```python
import holoviews as hv

colors = {
    'Training': 'blue',
    'Validation': 'red',
    'Extrapolation': 'green'
}

markers = {
    'Training': 'square',
    'Validation': 'diamond',
    'Extrapolation': 'circle'
}

scatter = hv.Overlay([
    hv.Scatter(
        df[df["Dataset"] == ds],
        kdims="n1",
        vdims=["n2", "MAPE"],
        label=ds
    ).opts(
        color=colors[ds],
        marker=markers[ds],
        size=10,
    )
    for ds in colors
])

labels = hv.Labels(df, kdims=["n1", "n2"], vdims=["MAPE"]).opts(
    text_font_size="10pt",
    yoffset=0.3
)

fig = (scatter * labels).opts(
    width=500,
    height=500,
    show_legend=True,
    legend_position="bottom_right",
    xticks=8,
    yticks=10,
    
)
```
