```python
import holoviews as hv
from holoviews import dim, opts
import pandas as pd
```

```python
hv.extension('plotly')
```

```python
df = pd.read_excel("./ideal_gas.xlsx",sheet_name="ideal_gas")
df.rename(columns={'index':'Dataset'},inplace=True)
```

```python
df['MAPE'] = df['perc']*100
```

```python
import plotly.express as px

fig = px.violin(
    df,
    x='Dataset',
    y='perc'
)

fig.update_layout(
    width=800,
    height=600,
    yaxis_title='MAPE'
)

fig.show()
```

```python
fig = px.violin(df, x='Dataset', y='MAPE', points='all')
fig.update_layout(template='simple_white')

fig.update_layout(
    xaxis_title="Dataset",
    yaxis_title="Relative Error (%)",
    showlegend=False,
    width=800
)

fig.show()
```

```python
import plotly.express as px

fig = px.scatter_3d(
    df,
    x='n',
    y='T',
    z='V',
    color='Dataset'
)

fig.update_layout(width=800, height=800,legend=dict(orientation="h", x=0.25,y=0.95))
fig.update_traces(marker=dict(size=5))
```

```python
import random
```

```python
points=[]
for _,row in df.loc[df['MAPE']>=50,['n','T','V','MAPE']].iterrows():
    points.append({"x": row['n'], "y": row['T'], "z": row['V'], "label": round(row['MAPE'],1)})
    
# Generate list of dict layouts
annotations_list = []
for p in points:
    annotations_list.append(dict(
        x=p["x"],
        y=p["y"],
        z=p["z"],
        text=p["label"],
        showarrow=True,
        xanchor='right',
        ax=random.choice([-20,-30,30,20]),
        ay=random.choice([-20,-30,30,20])
    ))

# Pass the completed list to the figure
fig.update_layout(scene=dict(annotations=annotations_list))
fig.show()
```
